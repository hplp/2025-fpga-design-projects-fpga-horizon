```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
image = cv2.imread('Roma.png', cv2.IMREAD_GRAYSCALE)

height, width = image.shape

kernel_size = 3  # Usually 3x3 for Sobel filter

```

```python
# Start timestamp
start_time = time.time()

# Apply Sobel operator (gradient in X and Y directions)
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=kernel_size)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=kernel_size)

# Compute magnitude of gradients (edge intensity)
sobel_magnitude = cv2.magnitude(sobel_x, sobel_y)

# Convert magnitude to uint8 for display
sobel_magnitude = np.uint8(np.absolute(sobel_magnitude))

# End timestamp
end_time = time.time()
elapsed_time = (end_time - start_time) * 1000 
```

```python
plt.figure(figsize=(25, 25))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sobel_magnitude, cmap='gray')
plt.title(f"Sobel Edge Detection (Kernel {kernel_size}x{kernel_size})")
plt.axis('off')
plt.show()
```

```python
print(f"Image size       : {width} x {height} pixels")
print(f"Kernel size      : {kernel_size}x{kernel_size}")
print(f"Sobel edge detection execution time: {elapsed_time:.3f} ms")
```
