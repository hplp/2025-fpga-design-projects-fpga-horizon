```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
# Load grayscale image
image = cv2.imread('Roma.png', cv2.IMREAD_GRAYSCALE)

height, width = image.shape


#Kernel size
r = 9
kernel_size = (r*2+1, r*2+1)
```

```python
# Start timestamp
start_time = time.time()

# Apply mean (box) blur using OpenCV
blurred_image = cv2.blur(image, kernel_size)
# End timestamp
end_time = time.time()
elapsed_time = (end_time - start_time) * 1000
```

```python
# Display original and blurred images side by side
plt.figure(figsize=(20, 20))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(blurred_image, cmap='gray')
plt.title(f"Box Blur ({kernel_size[0]}x{kernel_size[1]})")
plt.axis('off')

plt.show()
```

```python
print(f"Image size       : {width} x {height} pixels")
print(f"Kernel size      : {kernel_size[0]} x {kernel_size[1]}")

print(f"Box blur (mean filter) execution time: {elapsed_time:.3f} ms")
```
