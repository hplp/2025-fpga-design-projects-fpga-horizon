```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
# Load grayscale image
image = cv2.imread('Roma.png', cv2.IMREAD_GRAYSCALE)

height, width = image.shape

# Kernel size for Gaussian blur
r = 25  # Radius
kernel_size = (r*2+1, r*2+1)

# Standard deviation for Gaussian kernel (can be adjusted)
sigma = 5.0
```

```python
# Start timestamp
start_time = time.time()

# Apply Gaussian blur using OpenCV
gaussian_blurred_image = cv2.GaussianBlur(image, kernel_size, sigma)

# End timestamp
end_time = time.time()
elapsed_time = (end_time - start_time) * 1000  # Convert to milliseconds
```

```python
# Display the original and processed images side by side
plt.figure(figsize=(20, 20))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gaussian_blurred_image, cmap='gray')
plt.title(f"Gaussian Blur ({kernel_size[0]}x{kernel_size[1]}, σ={sigma})")
plt.axis('off')
plt.show()
```

```python
# Print metadata
print(f"Image size       : {width} x {height} pixels")
print(f"Kernel size      : {kernel_size[0]} x {kernel_size[1]}")
print(f"Gaussian blur (σ={sigma}) execution time: {elapsed_time:.3f} ms")
```
